```python
#Using Machine Learning to Predict the number of Goals Scored and Conceded in this year's World Cup.

from scipy import stats

Year = [2006,2010,2014,2018]
Goals_For = [12,8,18,14]

slope, intercept, r, p, std_err = stats.linregress(Year, Goals_For)

def myfunc(Year):
  return slope * Year + intercept

mymodel = list(map(myfunc, Year))

plt.scatter(Year, Goals_For)
plt.plot(Year, mymodel)
plt.show()
```

```python
#Importing more data sources to determine the number of goals scored by teams in the world cup qualification series between 2020 and 2022. The data sets are limited to European and South American teams as these make up the top 12 ranked teams in the world.

europe = pd.read_csv("international-wc-qualification-europe.csv")
```

```python
europe.head()
```

```python
south_america = pd.read_csv("international-wc-SA.csv")
south_america.head()
```

```python
#Concatenating the 2 datasets together.

wc_qualifications = pd.concat([europe[0:],south_america[0:]])
```

```python
print(wc_qualifications)
```

```python
#Sorting the data by the number of goals scored in descending values.

wc_qualifications = wc_qualifications.sort_values(by=['goals_scored'], ascending=False)
```

```python
print(wc_qualifications)
```

```python
#The south american teams played on average 7 more games in qualification than their european counterparts. We must derive the average goals scored and conceded per game to get some valuable insights.
#Ranking the teams based on their average goals scored per game.
```

```python
wc_qualifications["Avg_goals_scored"] = wc_qualifications["goals_scored"]/wc_qualifications["matches_played"]
```

```python
wc_qualifications = wc_qualifications.sort_values(by=['Avg_goals_scored'], ascending=False)
```

```python
wc_qualifications.head()
```

```python
#Ranking the teams based on their average goals conceded per game.

wc_qualifications["Avg_goals_conceded"] = wc_qualifications["goals_conceded"]/wc_qualifications["matches_played"]
```

```python
wc_qualifications = wc_qualifications.sort_values(by=['Avg_goals_conceded'], ascending=True)
```

```python
wc_qualifications.head()
```

```python
#Sorting the qualifying teams that meet the criteria of 1.85 or more goals scored per game average. This is the average goals per game scored in the tournament by the previous 4 winners. 

goals_tick = wc_qualifications.loc[wc_qualifications['Avg_goals_scored'] >= 1.857]
```

```python
print(goals_tick)
```

```python
#Sorting the qualifying teams that meet the criteria of 0.5 or less goals conceded per game average. This is the average goals per game conceded in the tournament by the previous 4 winners. 

defence_tick = wc_qualifications.loc[wc_qualifications['Avg_goals_conceded'] <= 0.5]
```

```python
print(defence_tick)
```

```python
#Performing an inner join on both DFs to get the teams that qualify for both goals scored and goals conceded conditions.

tipped_teams = goals_tick.merge(defence_tick, on='team_name')
```

```python
tipped_teams
```

```python
#Visualising the average goals scored and conceded throughout qualifying by the 7 teams that meet the criteria. As we can see from these plots, England scored the most goals on average and Switzerland conceded the least.


import seaborn as sns

Teams = ['Switzerland','Brazil','England','Denmark','France','Germany','Croatia']
Avg_Goals_Scored = [1.87,2.35,3.9,3.0,2.25,3.6,2.1]
Avg_Goals_Conceded = [0.25,0.294,0.3,0.3,0.375,0.4,0.4]

sns.barplot(x=Teams, y=Avg_Goals_Scored)
plt.show()

```

```python
sns.barplot(x=Teams, y=Avg_Goals_Conceded)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
x = Teams
y1 = Avg_Goals_Scored
y2 = Avg_Goals_Conceded

ax.plot(x,y1)
ax.plot(x,y2)

plt.show()
```
